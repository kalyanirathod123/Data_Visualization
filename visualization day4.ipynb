```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Automobile.csv')
```

```python
#plotting univariate distribution
#analysing of single column of data frame
#seaborn----distplot
```

```python
sns.distplot(data['highway_mpg'])
```

```python
sns.distplot(data['highway_mpg'])
plt.show() #removing the quote x-lable and y-lable
```

```python
sns.distplot(data['highway_mpg'],kde=False)
plt.show()
```

```python
sns.distplot(data['highway_mpg'],kde=False,rug=True)
plt.show()
```

```python
#plotting bivariate distribution
#jointplot
```

```python
sns.jointplot(data['engine_size'],data['horsepower'])
plt.show()
```

```python
sns.jointplot(data['engine_size'],data['horsepower'],kind='hex')
plt.show() #hexagon form of the graph
```

```python
#pairplot
sns.pairplot(data[['normalized_losses','engine_size','horsepower']])
plt.show()
```

```python
#plotting with categorical data
#strip plot
```

```python
sns.stripplot(data['fuel_type'],data['horsepower'],jitter=True)
plt.show()
```

```python
#boxplot
sns.boxplot(data['number_of_doors'],data['horsepower'])
plt.show()
```

```python
sns.barplot(data['body_style'],data['horsepower'],hue=data['fuel_type'])
plt.show()    #barplot
```

```python
#countplot
```

```python
sns.countplot(data['body_style'],hue=data['fuel_type'])
plt.show()
```
